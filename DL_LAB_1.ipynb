```python
import tensorflow as tf
import numpy as np

#Step-1: Convet data to numpy arrays and reshape for model input
x=np.array([1,2,3,4], dtype=float)# Reshape x to (num_samples, 1, 1)
y=np.array([2,4,6,8], dtype=float)

#Step-2: Define model using Input layer(recommended)
model=tf.keras.Sequential([
    tf.keras.Input(shape=(1,)), #preference over input_shape in Dense
    tf.keras.layers.Dense(1)
])

#Step-3: Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')

#Step-4: Train the model
model.fit(x,y,epochs=10)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#Step-1: Convet data to numpy arrays and reshape for model input
x=np.array([1,2,3,4], dtype=float)
y=np.array([2,4,6,8], dtype=float)

#Define the model
model=Sequential([
    Dense(1, input_shape=(1,))
])

#Compile the model
model.compile(optimizer='sgd', loss='mse')

#Train he model
model.fit(x,y,epochs=10)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

#Define the simple linear model
model = nn.Linear(1, 1) # Corrected: added out_features and changed '-' to '='

#Loss and optimizer
criterion=nn.MSELoss()
optimizer=optim.SGD(model.parameters(),lr=0.01)

#Dummy data
x=torch.tensor([[1.],[2.],[3.],[4.]])
y=torch.tensor([[2.],[4.],[6.],[8.]])

#Training loop
for epoch in range(10):
  optimizer.zero_grad()
  outputs=model(x)
  loss=criterion(outputs,y)
  loss.backward()
  optimizer.step()
  print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```
